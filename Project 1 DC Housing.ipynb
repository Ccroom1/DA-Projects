```python
#choosing columns to put in a new dataframe
DCdf = df[['PRICE' , 'QUADRANT' , 'ZIPCODE', 'ASSESSMENT_NBHD' , 'GBA' , 'BEDRM']]
GBAdf = DCdf[df['GBA'].notnull()]
PRICEdf = GBAdf[GBAdf['PRICE'].notnull()]
```

```python
PRICEdf.head()
```

```python
PRICEdf.describe()
```

```python
#Creating new catogories to group data in set
bins = [0, 1190, 1480, 1966, 45384]
group_names = ['Small' ,'Medium' ,'Large' ,'Huge' ]
PRICEdf['HomeSize'] = pd.cut(df['GBA'], bins, labels=group_names)
PRICEdf.head()
```

```python
PRICEdf.groupby(['QUADRANT','HomeSize']).mean()
```

```python
PRICEdf.groupby(['QUADRANT','HomeSize','ASSESSMENT_NBHD']).mean()
```

```python
MEANPRICE = PRICEdf['PRICE'].mean()
STDPRICE = PRICEdf['PRICE'].std()
toprange = MEANPRICE + STDPRICE * 1.96
botrange = MEANPRICE - STDPRICE * 1.96

copyPRICEdf = PRICEdf.copy()
copyPRICEdf = copyPRICEdf.drop(copyPRICEdf[copyPRICEdf['PRICE'] > toprange].index)
copyPRICEdf = copyPRICEdf.drop(copyPRICEdf[copyPRICEdf['PRICE'] < botrange].index)

copyPRICEdf.head()
```
